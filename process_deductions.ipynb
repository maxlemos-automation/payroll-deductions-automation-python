```python
import pandas as pd
import numpy as np
import csv
```

```python
fecha_inicio = "2024/01/01"
fecha_fin = "2024/01/15"
```

```python
adelantos = pd.read_excel("employee_deductions.xlsx")
codigoDescuentos = pd.read_excel("deduction_codes.xlsx")
```

```python
listaDescuentos = ["CI", "ANDAALQ", "ANDAPRE", "BROU", "BHU", "ANV", "AEOMM", "AEOMM P/C", "U.C.M.", "CASH", "COMAG"]
```

```python
# Replace NA values with 0 and select only the columns in listaDescuentos

adelantos =adelantos.fillna(0)[listaDescuentos]
```

```python
# Transforming the dataframe from wide to long format
adelantos = pd.melt(adelantos, 
        id_vars=["CI"],
        var_name="descuento",
        value_name="Importe")
```

```python
# Filter out rows where the "Importe" value is zero

adelantos = adelantos.loc[adelantos["Importe"] != 0, :]
```

```python
# Merging "adelantos" with "codigoDescuentos" to retrieve the codes

adelantos = pd.merge(adelantos, codigoDescuentos, 
                     how="left", 
                     left_on="descuento", 
                     right_on="descuento", 
                     left_index=False, 
                     right_index=False).sort_values("codDescuento")
```

```python
# Keep only the required columns

adelantos = adelantos[["CI", "codDescuento", "Importe"]]
```

```python
# Add start/end date and other required columns

adelantos["fechaInicio"] = fecha_inicio
adelantos["fechaFin"] = fecha_fin
adelantos["cantidad"] = 1
adelantos["cuotasInicio"] = 1
adelantos["cuotasFin"] = 1
adelantos["vaciaUno"] = ""
adelantos["vaciaDos"] = ""
```

```python
adelantos = adelantos[["vaciaUno", "CI", "vaciaDos", "fechaInicio", "codDescuento", "cantidad", "Importe", "cuotasInicio", "fechaFin", "cuotasFin"]]
```

```python
adelantos.to_csv("exportado.txt", sep=";", index=False, header=False)
```
